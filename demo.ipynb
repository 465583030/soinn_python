```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from numpy.random import normal
from soinn import Soinn

# generate data
n = 5000
sigma = 0.05
x = 10 * np.random.rand(n) - 5
data = np.array([[normal(x[i], sigma), normal(np.sin(x[i]), sigma)] for i in range(len(x))])

# initialize SOINN
s = Soinn(delete_node_period=1000)
# learn data in an online manner
for i in range(int(n)):
    s.input_signal(data[i, :])

# show SOINN's state
plt.plot(data[:, 0], data[:, 1], 'cx')
nodes = np.copy(s.nodes)
for k in s.adjacent_mat.keys():
    plt.plot(nodes[k, 0], nodes[k, 1], 'k')
plt.plot(nodes[:, 0], nodes[:, 1], 'ro')
```
